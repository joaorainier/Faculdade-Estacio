```python
# Import

import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
# Gráfico

class Grafico:
    def __init__(self, lista_gastos):
        self.lista_gastos = lista_gastos
        self.imprimir_graficos()

    def padrao_do_grafico(self):
        plt.xlabel('Dia')
        plt.ylabel('Valores (R$)')
        plt.title('Gráfico de Gastos')

    def imprimir_graficos(self):
        self.padrao_do_grafico()
        for gastos in self.lista_gastos:
            mLista = gastos.dicionario.items()
            cor = gastos.cor
            nome = gastos.nome
            x, y = zip(*mLista)
            plt.plot(x, y, label=nome, marker='o',
                     markerfacecolor='orange',
                     markersize=6,
                     color=cor,
                     linewidth=2)
        plt.legend()
        plt.show()

# Regressão Linear

    def regressao_linear(self, id_grafico):
        gastos = self.lista_gastos[id_grafico]
        mLista = gastos.dicionario.items()
        cor = gastos.cor
        nome = gastos.nome
        dias, valores = zip(*mLista)
        dias = np.array(dias)
        valores = np.array(valores)
        dias = dias.reshape(-1, 1)
        valores = valores.reshape(-1, 1)
        regr = LinearRegression()
        regr.fit(X=dias, y=valores)
        plt.plot(dias, regr.predict(dias),
                 color='pink',
                 label="Regressão Linear")

        x, y = zip(*mLista)
        plt.plot(x, y, label=nome,
                marker = 'o',
                markerfacecolor='yellow',
                markersize=6,
                color=cor,
                linewidth=2)

        plt.legend()
        plt.show()

```

```python
# Gastos

class Gastos:
    def __init__(self, dicionario, cor, nome):
        self.dicionario = dicionario
        self.cor = cor
        self.nome = nome

transporte = Gastos({1: 100, 5: 800, 17: 230, 20: 88, 29: 280}, 'darkblue', 'transporte')
alimentacao = Gastos({1: 60, 5: 769, 19: 311, 29: 280}, 'grey', 'alimentação')
vestuario = Gastos({1: 150, 5: 900, 16: 138, 20: 188, 30: 60}, 'green', 'vestuário')
lazer = Gastos({1: 120, 5: 654, 16: 128, 20: 242, 30: 42}, 'purple', 'lazer')
lista_gastos = [transporte, alimentacao, vestuario, lazer]
```

```python
#Saida

grafico = Grafico(lista_gastos)

id_mes = 3
grafico.regressao_linear(id_mes)

id_mes = 2
grafico.regressao_linear(id_mes)

id_mes = 1
grafico.regressao_linear(id_mes)

id_mes = 0
grafico.regressao_linear(id_mes)
```
